```python
x=df[['Hours Studied',
 'Previous Scores',
 'Sleep Hours',
 'Sample Question Papers Practiced', ]]
y=df['Performance Index']
```

```python
X=sm.add_constant(x)
mo=sm.OLS(y,X)
result=mo.fit()
print(result.summary())
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2,random_state=5)
```

```python
X_train=sm.add_constant(x_train)
mo1=sm.OLS(y_train,X_train)
result1=mo1.fit()
print(result1.summary())
```

```python
X_test=sm.add_constant(x_test)
y_pred2 =result1.predict(X_test)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)
```

```python
y_pred_series=pd.Series(y_pred, name='y_pred')
y_pred2_series=pd.Series(y_pred2, name='y_pred2')
y_test_series=pd.Series(y_test, name='y_test')

y_pred_series.reset_index(drop=True, inplace=True)
y_pred2_series.reset_index(drop=True, inplace=True)
y_test_series.reset_index(drop=True, inplace=True)

combined=pd.concat([y_pred_series,y_pred2_series,y_test_series], axis=1)
combined.head(30)
```
